# Downloading dataset

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("imakash3011/customer-personality-analysis")

print("Path to dataset files:", path)
```

# Importing necessary libraries

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading dataset to a dataframe
df = pd.read_csv(path + "/marketing_campaign.csv", sep="\t")
# Temporarily show all columns
with pd.option_context('display.max_columns', None):
    print(df.head(20))
```

```python
# Null values
df.isnull().sum()
```

```python
df['Income'].isnull().sum()
```

```python
# As income is a numerical data let's fill the null values with median value
df['Income'].fillna(df['Income'].median(), inplace=True)
```

```python
# Identify duplicate rows
df.duplicated().sum()
```

```python
# Remove duplicate rows (In our case we don't have any)
df.drop_duplicates(inplace=True)
```

```python
# Standardize and clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# check unique values in Education column
df['Education'].unique()
```

```python
# count of each unique values
df['Education'].value_counts()
```

```python
### Values are in correct form in each column so we do not need to modify them.
```

```python
# Convert Dt_customer column to date-time type.
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True)
```

```python
# Rename column headers to be clean and uniform
df.columns = df.columns.str.strip() \
                       .str.lower() \
                       .str.replace(' ', '_') \
                       .str.replace('[^0-9a-zA-Z_]', '', regex=True)
```

```python
df.head()
```

```python
# check for unique values in marital_status column
df['marital_status'].unique()
```

```python
# Replace absurd words so that the column only contain Single, Married and Divorced and Widow entry and nothing else.
marital_status_map = {
    'Single': 'Single',
    'Married': 'Married',
    'Together': 'Married',
    'Divorced': 'Divorced',
    'Widow': 'Widow',
    'Alone': 'Single',
    'Absurd': 'Single',
    'YOLO': 'Single'
}

df['marital_status'] = df['marital_status'].replace(marital_status_map)
```
